```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import numpy as np

data={'title':['Death Note',
               "The Ancient Magus' Bride",
               'Snow White with Red Hair',
               'Pretty Rhythm',
               'Orange',
               "Why Did Raeliana End up At The Duke's Mansion",
               "Saint's Magic is Omnipotent"],
      'genres':["Psychological thriller|horror|mystery",
                "Adventure|fantasy|magic",
                "magic|shoujo",
                "Idol|music|Shoujo",
                "Isekai|thriller|mystery",
                "fantasy|magic",
                "fantasy|magic|shoujo"]}

df=pd.DataFrame(data)

df['genre_list'] = df['genres'].apply(lambda x: x.split('|'))
all_genres=list(set([genre for sublist in df['genre_list'] for genre in sublist]))
for genre in all_genres:
  df[genre]=df['genre_list'].apply(lambda x: int(genre in x))
X=df[all_genres]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
kmeans=KMeans(n_clusters=3,random_state=42)
df['cluster']=kmeans.fit_predict(X_scaled)

def recommend_anime(anime_title,df):
  if anime_title not in df['title'].values:
    return "Anime not found in the dataset."

  cluster=df[df['title']==anime_title]['cluster'].values[0]
  recommended=df[df['cluster']==cluster]['title']
  return recommended.values.tolist()

print(recommend_anime('Snow White with Red Hair',df))
```
